```python
import pandas
credit1 = pandas.read_excel('default of credit card clients-NotNormalized.xlsx')
credit2 = pandas.read_excel('default of credit card clients-Normalized by Credit limit.xlsx')

credit = credit1
```

```python
#Gender
male = 0
female = 0
for i in credit['SEX']:
    if i == 1:
        male += 1
    elif i == 2:
        female += 1
print('# of males / Probability',male,'/',round(male/credit['SEX'].count(),3),
      '\n# of females ',female,'/',round(female/credit['SEX'].count(),3))
```

```python
import matplotlib.pyplot as plt
gender = [male,female]
plt.pie(gender,labels=['male','female'],autopct='%.2f')
plt.show()
```

```python
#education
zero = 0
graduate = 0
bachelor = 0
highschool = 0
other = 0
five = 0
six = 0

for i in credit['EDUCATION']:
    if i == 1:
        graduate += 1
    elif i == 2:
        bachelor += 1
    elif i == 3:
        highschool += 1
    elif i == 4:
        other += 1
    elif i == 5:
        five += 1
    elif i == 6:
        six += 1
    elif i == 0:
        zero += 1

print('\n# with graduate education / Probability',graduate,'/',round(graduate/credit['EDUCATION'].count(),3),
      '\n# with bachelor education / Probability',bachelor,'/',round(bachelor/credit['EDUCATION'].count(),3),
      '\n# with highschool education / Probability',highschool,'/',round(highschool/credit['EDUCATION'].count(),3),
      '\n# with other education / Probability',other,'/',round(other/credit['EDUCATION'].count(),3),
      '\n# with five education / Probability',five,'/',round(five/credit['EDUCATION'].count(),3),
      '\n# with six education / Probability',six,'/',round(six/credit['EDUCATION'].count(),3),
      '\n# with zero education / Probability',zero,'/',round(zero/credit['EDUCATION'].count(),3),
      '\nCount',credit['EDUCATION'].count())
```

```python
education = (highschool,six,graduate,other,bachelor,five) 
Education = ('highschool','six','graduate','other','bachelor','five') #categories
plt.pie(education,labels=Education,autopct='%.2f')
plt.show()
```

```python
#marital status
married = 0
single = 0
others = 0
zero = 0
for i in credit['MARRIAGE']:
    if i == 1:
        married += 1
    elif i == 2:
        single += 1
    elif i == 3:
        others += 1
    elif i == 0:
        zero += 1
print('\n# married / Probability',married,'/',round(married/credit['MARRIAGE'].count(),3),
      '\n# single / Probability',single,'/',round(single/credit['MARRIAGE'].count(),3),
      '\n# others / Probability',others,'/',round(others/credit['MARRIAGE'].count(),3),
      '\n# zero / Probability',zero,'/',round(zero/credit['MARRIAGE'].count(),3))
```

```python
marriage = (married,others,single,zero) 
Marriage = ('married','others','single','zero') #categories
plt.pie(marriage,labels=Marriage,autopct='%.2f')
plt.show()
```
